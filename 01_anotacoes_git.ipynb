# Introdução ao Git e ao Github

***Security Hash Algorithm*** (SHA): é um conjunto de funções _hash_ criptográficas projetadas pela _NSA_. Em termos práticos a função disso é criar um identificador de quarenta dígitos para cada arquivo.  

#### Objetos fundamentais do Git:
    
- Blob
- Tree
- Commit

## Alguns comandos iniciais

- `git init` - Iniciar repositório.
- `git add` - Adiciona o arquivo ao repositório.
- `git commit` - Cria um _commit_.
- `git status` - Cria um resumo do status de cada arquivo do repositório em questão em relação ao ciclo do vida.
- `git log` - Dá informações sobre os commits realizados.
- `git restore 'nome_do_arquivo'` - Descarta as mudanças realizadas e retorna o arquivo do commit anterior.
- `git restore --staged 'nome do arquivo'` - Retira o arquivo da Stage Area

## Opções de comando para adicionar realmente TODOS os arquivos modificados para a área de staging

- `git add --all`

- `git add -A`

- `git add .`

## Configurar informações do autor

- `git config --global user.name "George Pontes"`
- `git config --global user.email george.pontes02@gmail.com`

# Branches

- `git branch` - Ver lista de ramos (_branches_);
- `git branch nome_do_branch` - Criar um _branch_;
- `git branch -m nome_atual novo_nome` - Renomear _branch_;
- `git branch -D nome_do_branch` - Deletar um _branch_;
- `git checkout` - Sair do branch atual;
- `git checkout main` - Sair de um determinado _branch_ e entrar no _branch_ main;
- `git merge nome_do_branch` - Combina a sequência de commits do ramo atual com o de outro ramo, em um histórico unificado.

## Repositórios Remotos



### Adicionar repositórios remotos através de seus respectivos links (Geralmente o nome dos links de repositório vai ser `origin`.)


`git remote add NomeDoLink LinkEmSi.com` 

### Ver quais repositórios remotos estão cadastrados

`git remote -v`

### Empurrar repositório local para o remoto

`git push NomeDoLink master`

### Resolução de conflitos _merge_ (entre o repositório local e o remoto)

1. Usar `git pull NomeDoLink master` para baixar a versão mais atual do repositório.
2. Editar os arquivos para resolver o conflito.
3. `git add *` para adicionar tudo o que foi modificado. 
4. `git commit -m "mensagem de sua preferência"` 
5. `git push NomeDoLink master`para empurrar a versão com o conflito resolvido.

Resumindo: Baixa o repositório, resolve o conflito e Empurra ele de volta para o repositório remoto.

### Clonar repositório remoto

`git clone linkemsi.com`
