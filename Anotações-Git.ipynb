#### Alguns comandos iniciais

- `git init` - Iniciar repositório.
- `git add` - Adiciona o arquivo ao repositório.
- `git commit` - Cria um _commit_.
- `git status` - Cria um resumo do status de cada arquivo do repositório em questão em relação ao ciclo do vida.


### Adicionar repositórios remotos através de seus respectivos links

`git remote add NomeDoLink LinkEmSi.com` 

### Ver quais repositórios remotos estão cadastrados

`git remote -v`

### Empurrar repositório local para o remoto

`git push NomeDoLink master`

### Resolução de conflitos _merge_ (entre o repositório local e o remoto)

1. Usar `git pull NomeDoLink master` para baixar a versão mais atual do repositório.
2. Editar os arquivos para resolver o conflito.
3. `git add *` para adicionar tudo o que foi modificado. 
4. `git commit -m "mensagem de sua preferência"` 
5. `git push NomeDoLink master`para empurrar a versão com o conflito resolvido.

Resumindo: Baixa o repositório, resolve o conflito e Empurra ele de volta para o repositório remoto.

### Clonar repositório remoto

`git clone linkemsi.com`
